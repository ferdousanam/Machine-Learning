```python
# Load Data From CSV
raw_data = pd.read_csv('datasets/kddcup.data_10_percent.gz', header=None)
```

```python
# Conditions start to be in categories
DOS = (labels[41] == 'back.') | (labels[41] == 'land.') | (labels[41] == 'neptune.') | (labels[41] == 'pod.') | (labels[41] == 'smurf.') | (labels[41] == 'teardrop.')
U2R = (labels[41] == 'buffer_overflow.') | (labels[41] == 'loadmodule.') | (labels[41] == 'perl.') | (labels[41] == 'rootkit.')
R2L = (labels[41] == 'ftp_write.') | (labels[41] == 'guess_passwd.') | (labels[41] == 'imap.') | (labels[41] == 'multihop.') | (labels[41] == 'phf.') | (labels[41] == 'spy.') | (labels[41] == 'warezclient.') | (labels[41] == 'warezmaster.')
probe = (labels[41] == 'satan.') | (labels[41] == 'ipsweep.') | (labels[41] == 'portsweep.') | (labels[41] == 'nmap.')
# Conditions end

labels['categories'] = np.where(DOS, 'dos', np.where(U2R, 'u2r', np.where(R2L, 'r2l', np.where(probe, 'probe', labels[41]))))

# This is just for testing purpose
# labels.loc[labels['categories'].isin(['u2r'])]
```
